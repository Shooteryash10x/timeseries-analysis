```python
# Choose columns to plot (replace with your desired column names)
x_column = 'Date'  
y_column = 'Close'

# Create the plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=x_column, y=y_column, data= df)
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title(f'{y_column} over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
window_size = 10  # Adjust the window size as needed
df['SMA'] = df['Close'].rolling(window=window_size).mean()

# Extend the index for forecasting
# Assuming 'Date' column contains dates in 'YYYY-MM-DD' format
# Convert 'Date' column to datetime objects if it's not already
 # Assuming 'Date' column contains dates in 'DD-MM-YYYY' format
# Convert 'Date' column to datetime objects if it's not already
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

last_date = df['Date'].iloc[-1]  # Get the last date from the 'Date' column
future_dates = [last_date + pd.DateOffset(days=i) for i in range(1, 31)]
future_dates_str = [date.strftime('%Y-%m-%d') for date in future_dates]  # Format as 'YYYY-MM-DD'

# Forecast using the last valid SMA value
# Instead of using df['SMA'][-1], use df['SMA'].iloc[-1] to access the last valid value
forecast = [df['SMA'].iloc[-1]] * 30  

# Create a DataFrame for the forecast
forecast_df = pd.DataFrame({'SMA': forecast}, index=future_dates)  # Use datetime objects for index

# Combine the original df with the forecast
combined_df = pd.concat([df, forecast_df])


# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(combined_df['Date'], combined_df['Close'], label='Actual')  # Use 'Date' column for x-axis
plt.plot(combined_df['Date'], combined_df['SMA'], label='SMA')  # Use 'Date' column for x-axis
plt.plot(forecast_df.index, forecast_df['SMA'], label='Forecast', linestyle='--')  # Highlight the forecast
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Stock Price Forecast using Simple Moving Average')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
