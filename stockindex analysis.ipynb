```python
#time series analysis and modelling
### time series data is type of data which is associated with date and time 
### it can be hourly, monthly,yearly the only condition is that interval has to be same 
### it has property that if some pattern exist in past it is lightly that that pattern will repeat in future eg.stock market,weather
```

```python
index dataset consist of index movement of 4 diff. countries from 1994 to 2018
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("index (1).csv")
```

```python
df
```

```python
df1=df.copy()
```

```python
df1
```

```python
df1.head()
```

```python
df1.describe()
```

```python
df1.spx.plot(figsize=(20,5), title="standar & p500 prices")
plt.xlabel("day", fontsize=14)
plt.ylabel("index value", fontsize=14)
plt.show()
```

```python
df1.ftse.plot(figsize=(20,5), title="ftse100 prices")
plt.xlabel("day", fontsize=14)
plt.ylabel("index value", fontsize=14)
plt.show()
```

```python
df1.spx.plot(figsize=(20,5), title="standar & p500 prices")
df1.ftse.plot(figsize=(20,5), title="ftse100 prices")
plt.title("s&p vs ftse")
plt.show()
```

```python
import scipy.stats
import pylab
```

```python
scipy.stats.probplot(df1.spx, plot= pylab)
plt.title("qq plot", size=24)
pylab.show()
```

```python
scipy.stats.probplot(df1.ftse, plot= pylab)
plt.title("qq plot", size=24)
pylab.show()
```

```python
scipy.stats.probplot(df1.dax, plot= pylab)
plt.title("qq plot", size=24)
pylab.show()
```

```python
scipy.stats.probplot(df1.nikkei, plot= pylab)
plt.title("qq plot", size=24)
pylab.show()
```
